```python
# jetzt legen wir zuerst eine neue, leere Liste an, die wir mit den Daten in form eines dictionarys füllen:
master_df_list = []

# dann holen wir uns jene beiden daten, die für jede Zeile geich sein sollen:
datum = df_test.loc[2, 2]
geschaeft = df_test.loc[0, 1]

# Achtung: bei beiden muss man später noch einen Test einbauen, falls das je nach Tabelle nicht genau an diesem Ort ist!

zeilen = len(df_test.index)

for zeile in range(1,zeilen):
    if len(df_test.loc[zeile, 1]) >=1:
        name = df_test.loc[zeile, 0]
        kanton = df_test.loc[zeile, 3]
        verhalten = df_test.loc[zeile, 2]

        mini_dict = {
            "Datum" : datum,
            "Geschäft" : geschaeft,
            "Politiker" : name,
            "Kanton" : kanton,
            "Verhalten" : verhalten}
        
        master_df_list.append(mini_dict)
        
    if len(df_test.loc[zeile, 4]) >=1:
        name = df_test.loc[zeile, 4]
        kanton = df_test.loc[zeile, 7]
        verhalten = df_test.loc[zeile, 6]

        mini_dict = {
            "Datum" : datum,
            "Geschäft" : geschaeft,
            "Politiker" : name,
            "Kanton" : kanton,
            "Verhalten" : verhalten}
        
        master_df_list.append(mini_dict)
```
