```python
import pandas as pd
```

```python
path = "COVID19VaccDosesAdministered_AKL10_w.csv"
```

```python
df_impf1 = pd.read_csv(path)
```

## Nur jüngstes Datum

```python
df_Mai = df_impf1[df_impf1["date"] == 202117]
df_Mai
```

```python
#grosse Kantone rausfiltern
df_MaiGross = df_Mai[(df_Mai["geoRegion"] == "ZH") | (df_Mai["geoRegion"] == "VD") | (df_Mai["geoRegion"] == "BE")]
```

```python
df_Mai.columns
```

```python
# weglöschen der unnötigen Zahlen
df_MaiGross = df_MaiGross.drop(['type', 'version', 'granularity',
       'freq', 'prct'], axis = 1)
```

```python
df_MaiGross
```

```python
#Seit Kampagnenanfang
df_pivot = df_MaiGross.pivot(index='altersklasse_covid19', columns = "geoRegion", values = "per100PersonsTotal")
df_pivot
```

```python
df_pivot2 = df_MaiGross.pivot(index='altersklasse_covid19', columns = "geoRegion", values = "per100Persons")
df_pivot2
```
