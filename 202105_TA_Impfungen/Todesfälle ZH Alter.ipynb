```python
import pandas as pd
```

```python
path = "COVID19Death_geoRegion_AKL10_w.csv"
```

```python
df_impf1 = pd.read_csv(path)
```

## Nur j√ºngstes Datum

```python
df_Mai = df_impf1[df_impf1["datum"] == 202117]
df_Mai
```

```python
#grosse Kantone rausfiltern
df_MaiGross = df_Mai[(df_Mai["geoRegion"] == "ZH") | (df_Mai["geoRegion"] == "VD") | (df_Mai["geoRegion"] == "BE")]
```

```python
df_Mai.columns
```

```python
# wegl√∂schen der unn√∂tigen Zahlen
df_MaiGross = df_MaiGross.drop(["entries",'freq', 'prct', 'inz_entries', 'inzsumTotal', 'type',
       'type_variant', 'version', 'datum_unit', 'datum_dboardformated',
       'entries_diff_abs', 'entries_diff_inz', 'entries_diff_pct', 'prct_diff',
       'timeframe_all', 'timeframe_phase2', 'timeframe_phase2b',
       'timeframe_phase3', 'timeframe_28d', 'timeframe_14d'], axis = 1)
```

```python
df_MaiGross
```

```python
# Hier l√∂sche ich die NaN-Zeilen raus, wegen denen die Rechnung sonst nicht geht.
df_MaiGross.dropna(inplace = True)
```

```python
df_MaiGross["pop"] = df_MaiGross["pop"].round(0).astype(int, inplace = True)
```

```python
df_MaiGross.dtypes
```

```python
df_MaiGross["Anteilsm√§ssig"] = 10000000 / ((df_MaiGross["pop"]) * df_MaiGross["sumTotal"])
```

```python
df_MaiGross
```

```python
#Seit Kampagnenanfang
df_pivot = df_MaiGross.pivot(index='altersklasse_covid19', columns = "geoRegion", values = "Anteilsm√§ssig")
df_pivot
```

```python
import numpy as np 
import seaborn as sns
%matplotlib inline

sns.heatmap(df_pivot, annot=True, cmap="Blues")
```
